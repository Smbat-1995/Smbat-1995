# __Variables__

Variables
One of the most basic and powerful concepts is that of a __variable__.
A variable assigns a name to a value.

Always preferred to use variables over values:
    
- Easier to update code
- Easier to understand code (useful naming)

__Objects__

A object is the fundamental thing that a program manipulates. Values can be 
- “Hello, world!”, 42, 12.34, True

There is no need to declare the type of the value/object. Python dynamicly assigns the type to the value/object.
Values have types...

Objects can be:
- scalar (cannot be subdivided)
- non-scalar (have internal structure that can be assessed)

Scalar Data types

- Floating(Float)
- Integer
- bool
- NoneType

__Tips__

- Variable names can contain only letters, numbers, and underscores

- Avoid using Python keywords and function names as variable names

| Operator | Meaning | Example | Algrebraic|

__Comparison Operators__

| Operator | Meaning | Example | 
| ---|------|------|
| > | Greater that | x > y | 
| < | Less that | x < y | 
| == | Equal to | x==y | 
| !=| Not equal to | x!=y | 
| >= | Greater than or equal to | x>=y | 
| <= | Less than or equal to | x<=y |

__Asignment operators__

| Operator | Example | Equivalent | 
| ---|------|------|
| =  | x = 5 | x = 5 | 
| += | x += 5 | x = x + 5 | 
| -= | x -= 5 | x = x - 5 |

| %= | x %= 5 | x = x % 5 |
| //= | x //= 5 | x = x // 5 |
| **= | x **= 5 | x = x ** 5 |


__Logical operators__

| Operator | Meaning | Example | 
| ---|------|------|
| and | True if both the operands are true | x and y | 
| or  | True if either of the operands is true | x or y | 
| not | True if operand is false | not x | 


__Membership operators__

| Operator | Meaning | Example | 
| ---|------|------|
| in | True if value/variable is found in the sequence | 5 in x | 
| not in | True if value/variable is not found in the sequence | 5 not in x |
