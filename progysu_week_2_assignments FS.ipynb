__Exercise 17*: Square Root__

Write a program that implements Newton’s method to compute and display the square root of a number entered by the user. The algorithm for Newton’s method follows:

    Read x from the user 
        Initialize guess to x/2
        While guess is not good enough
               Update guess to be the average of guess and x/guess
    
When this algorithm completes, guess contains an approximation of the square root. The quality of the approximation depends on how you deﬁne “good enough”. Guess is considered good enough when the absolute value of the difference between guess∗guess and x is less than or equal to 10^−12.


__Exercise 18*: Simple Encryption__

You get a message which is a number and you have to encrypt it.The idea behind this encryption is simple (and as a result, it provides no protection against modern code breaking techniques). Each integer in the original number is shifted by 3 places. As a result, 1 becomes 4, 2 becomes 5, 3 becomes 6, 4 becomes 7, etc. The last integers  are wrapped around to the beginning: 7 becomes 0, 8 becomes 2 and 9 becomes 3.

1.  Write a program that implements the encryption. 
2. Allow the user to supply the message and the shift amount, and then display the shifted message.
3. Also support negative shift values so that it can be used both to encode messages and decode messages.
