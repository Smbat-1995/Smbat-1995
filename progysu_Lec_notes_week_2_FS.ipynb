<font size = 3 color = Red>Control Flow: WHILE <font>

```python
while <condition>:
    <expression>
    <expression>
    ------
```

In case of __while__ loop:

* __conditon__ evaluates to a Boolean
* if __condition__ is True, do all the steps inside the while block
* check __condition__ again
* repeat until __condition__ is False

```python
n = input("You're in the lost Forest. Go left or right?")
while n == 'right':
    n = input("You're in the lost Forest. Go left or right?")
print('You got out of the Lost Forest.')
```

__How to iterate through numbers in a sequence?__

```python
n = 0
while n<5:
    print(n)
```

```python
# more complicated with while loops
n = 0
while n<5:
    print(n)
    n += 1
```

```python
# shortcut with for loop
for n in range(5):
    print(n)
```

In case of __for__ loop:
* each time through the loop, __variable__ takes a value

* next time, __variable__ gets the prev value + 1

<font size = 3 color = Red> Continue Statement <font>

* immediately stops the iteration
* skips remaining expressions in code block
* continues with the next iteration of the smallest enclosing loop.

### For VS while LOOPS

__for__ loops 

* know number of iterations
* can end early via break
* uses a counter 
* can rewrite a for loop using while loop

__while__ loops

* unbounded number of iterations
* can end early via break
* can use a counter but must initialize before loop and increment it inside loop
* may not be able to rewrite a while loop using a for loop

__Task Find the cube root of an integer:__

```python
cube = int(input('Please provide a number'))
for guess in range(cube+1):
    if guess**3 == cube:
        print("Cube root of", cube, "is", guess)
        break
else:
    print('No cube root for this number')
```

__What if the number is negative?__

```python
cube = 27
num_guesses = 0
for guess in range(abs(cube)+1):
    num_guesses += 1
    if guess**3 >= abs(cube):
        break
if guess**3 != abs(cube):
    print(cube, 'is not a perfect cube')
else:
    if cube < 0:
        guess = -guess
    print(num_guesses)
    print('Cube root of '+str(cube)+' is '+str(guess))
```

__Final form__

```python
ans = 0
neg_flag = False
x = int(input("Enter an integer: "))
if x < 0:
    neg_flag = True
while ans**3 < x:
    ans = ans + 1
if ans**3 == x:
    print("Square root of", x, "is", ans)
else:
    print(x, "is not a perfect square")
    if neg_flag:
        print("Just checking... did you mean", -x, "?")
```

### Guest and check

* the process above also called exhaustive enumeration
* given a problem...

* you are able to check if the solution is correct

### Approximate solutions


* good enough solution
* start with a guess and increment by some small value
* keepguessing if |guess^3-cube| >= epsilon for some small epsilon

```python
cube = float(input())
epsilon = 0.01
guess = 0.0
increment = 0.0001
num_guesses = 0
while abs(guess**3 - cube) >= epsilon and guess <= cube :
    guess += increment
    num_guesses += 1
print('num_guesses =', num_guesses)
if abs(guess**3 - cube) >= epsilon:
    print('Failed on cube root of', cube)
else:
    print(guess, 'is close to the cube root of', cube)
```
