### Approximate solutions


* good enough solution
* start with a guess and increment by some small value
* keepguessing if |guess^3-cube| >= epsilon for some small epsilon

```python
cube = float(input())
epsilon = 0.01
guess = 0.0
increment = 0.0001
num_guesses = 0
while abs(guess**3 - cube) >= epsilon and guess <= cube :
    guess += increment
    num_guesses += 1
print('num_guesses =', num_guesses)
if abs(guess**3 - cube) >= epsilon:
    print('Failed on cube root of', cube)
else:
    print(guess, 'is close to the cube root of', cube)
```

### Bisection Search

* half interval each iteration
* new guess is halfway in between 
* to illustrate, letâ€™s play a game!

```python
cube = float(input(''))
epsilon = 0.0000001
num_guesses = 0
low = 0
high = cube
guess = (high + low)/2.0
while abs(guess**3 - cube) >= epsilon:
    if guess**3 < cube :
        low = guess
    else:
        high = guess
    guess = (high + low)/2.0
    num_guesses += 1
print ('num_guesses =', num_guesses)
print (guess, 'is close to the cube root of', cube)
```

__Default Arguments__

It is sometimes convenient to have default arguments.

```python
def factorial(x):
    """ Calculates the factorial of the number"""
    if x == 0:
        print (1)
    elif x < 0:
        print('Error')
    else: 
        i = 1
        fact = 1
        while i <= x:
            fact *= i
            i += 1
        print (fact)
            
```

```python
factorial(3)
```

__Passing by value__

Parameters to functions are references to objects, which are passed by value. When you pass a variable to a
function, python passes the reference to the object to which the variable refers (the value). Not the variable
itself.

If the value passed in a function is immutable, the function does not modify the callerâ€™s variable. If the value
is mutable, the function may modify the callerâ€™s variable in-place:

```python
def modif(x, y, z):
    x = 23
    y.append(42)
    z = [99]
    print(x)
    print(y)
    print(z)
```

```python
a = 77
b = [99]
c = [28]
modif(a,b,c)
```

```python
print(a)
```

```python
print(b)
```

```python
print(c)
```
